```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, time, date
%matplotlib inline
```

```python
df = pd.read_csv('clima.csv')
```

```python
df['FECHA'] = pd.to_datetime(df['FECHA'])
```

```python
del df['ID_ESTACION.1']
df['ANO'] = df['FECHA'].apply(lambda x: x.year)
```

```python
df.head()
```

```python
df_heatmap = pd.core.frame.DataFrame({'COUNT' : df.reset_index().groupby('NOMBRE')['ANO'].value_counts()}).reset_index()
plt.figure(figsize=(20,40))
plt.title('Total')
sns.heatmap(df_heatmap.pivot_table(values='COUNT',index='NOMBRE',columns='ANO'),cmap='coolwarm',linecolor='white',linewidths=1,annot=True,fmt='g')
```
